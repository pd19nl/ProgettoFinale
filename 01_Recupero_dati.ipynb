```python
# =============================================================
# =============================================================
#
# # librerie di supporto 
import numpy as np
import pandas as pd


# =============================================================
# =============================================================
#
#librerie fase recupero dati
import requests as api
```

```python
# =============================================================
# =============================================================
#
# recupero dati incidenti
#
# =============================================================
# =============================================================


#url dati incidenti
urldati_incidenti="https://esploradati.istat.it/SDMXWS/rest/data/41_983"

#NB aggiungo nella richiesta di ottenere il body della risposta
#headers_incidenti={'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}
headers_incidenti={'Accept': 'application/json'}
rq_incidente = api.get(urldati_incidenti, headers=headers_incidenti)
```

```python
#risposta http
if (rq_incidente.status_code==200):
    print(f"Risposta http success ({rq_incidente.status_code})")
else:
    print(f"Errore risposta http: {rq_incidente.status_code}" )
```
