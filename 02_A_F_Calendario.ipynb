```python
#scelta di fare il calendario da codice per imparare ad usare python
# infatti per fare il codice ho impegato tanto tempo
datainizio=f'{str(pd_config.GetAnnoInizioAnalisi())}-01-01'
datafine=f'{str(pd_config.GetAnnoInizioAnalisi() + pd_config.GetNrAnniAnalisi())}-12-31'

print(f'datainizio: {datainizio} - datafine: {datafine}' )
#crea un dataframe con 1 colonna con il range di date da a
dfcalendario = pd.DataFrame({"Data": pd.date_range(datainizio, datafine)})
```

```python
#aggiungo colonna indice
dfcalendario = dfcalendario.reset_index()
```

```python
dfcalendario.head(5)
```

```python
#aggiungo le colonne legate al tempo per le analisi e filtri
dfcalendario["Anno"] = dfcalendario.Data.dt.year
dfcalendario["Mese"] = dfcalendario.Data.dt.month
dfcalendario["Giorno"] = dfcalendario.Data.dt.day
#dfcalendario["GG_name"] = dfcalendario.Data.dt.day_name
#0/1/2
# get the day of the week (Note: Monday is coded as 0, and Sunday as 6)
dfcalendario["GGSettimana"] = dfcalendario.Data.dt.day_of_week #STESSO VALORE DI dfcalendario.Data.dt.weekday
#il nome del giorno della settimana
dfcalendario["GGSettimanaDsc"] = "LUN"


#non va dfcalendario["MeseDesc"] = dfcalendario.Data.dt.month_name
dfcalendario["GGDelMese"] = dfcalendario.Data.dt.days_in_month
```

```python
#aggiorna il datase --> fa schivo ma si migliora quando si ha tempo
def AssegnaGiorno(giornoSettimana:int,nomeDelGiorno:str):
  MASCHERA_DATI = dfcalendario["GGSettimana"]==giornoSettimana
  dfcalendario.loc[MASCHERA_DATI,"GGSettimanaDsc"]=nomeDelGiorno
```
