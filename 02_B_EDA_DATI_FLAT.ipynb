```python
df_regioni_an = pd.read_csv( pd_common.GetFolderAnalisi()+ "//ElencoRegioni.csv")
```

```python
df_ripgeo_an = pd.read_csv( pd_common.GetFolderAnalisi()+ "//ElencoRipartizioniGeografiche.csv")
```

```python
#merge tra regioni e ripartizione geografica
# la chiave di relazione è COD_RIP per ambedue le tb
df_reg_ripgeo_merge = df_regioni_an.merge(df_ripgeo_an[['COD_RIP','DEN_RIP']],
                                            how='inner', 
                                            left_on='COD_RIP', 
                                            right_on='COD_RIP')
```

```python
df_reg_ripgeo_merge.columns
```

```python
df_reg_ripgeo_merge.head(3)
```

```python
nr_rec_reg = df_regioni_an.COD_REG.count()
nr_rec_merge_reg = df_reg_ripgeo_merge.COD_REG.count()
print(f'Sorgente: {nr_rec_reg} - Mege: {nr_rec_merge_reg}')
```

```python
df_comuni_an.columns
```

```python
#merge tra comuni e regioni-ripartizione geografica
# la chiave di relazione è COD_REG per ambedue le tb
df_com_reg_geo_merge = df_comuni_an.merge(df_reg_ripgeo_merge[['anno_riferimento', 'COD_REG', 'DEN_REG','COD_RIP','DEN_RIP']],
                                            how='inner', 
                                            left_on=['anno_riferimento','COD_REG'], 
                                            right_on=['anno_riferimento','COD_REG'])
```

```python
df_com_reg_geo_merge.columns
```

```python
nr_rec_com = df_comuni_an.PRO_COM.count()
nr_rec_merge_com = df_com_reg_geo_merge.PRO_COM.count()
print(f'Sorgente: {nr_rec_com} - Mege: {nr_rec_merge_com}')
```

```python
#SALVO IL DATASET FINALE

pd_common.SalvaDataset(df_com_reg_geo_merge, f"ElencoComuniRegioniRipGeografiche",False)
```
